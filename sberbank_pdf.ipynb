```python
page1 = doc.loadPage(0)
page1text = page1.getText("text")
```

```python
parse_list = [
    'Гематокрит',
    'Гемоглобин',
    'Эритроциты',
    'MCV (ср. объем эритр.)',
    'RDW (шир. распред. эритр) 11.9',
    'MCH (ср. содер. Hb в эр.)',
    'МСHС (ср. конц. Hb в эр.)',
    'Тромбоциты',
    'Лейкоциты', 
    'Нейтрофилы (общ.число), ',
    'Лимфоциты, %',
    'Моноциты, %',
    'Эозинофилы, %',
    'Базофилы, %',
    'Нейтрофилы, абс.',
    'Лимфоциты, абс.', 
    'Моноциты, абс.',
    'Эозинофилы, абс.',
    'Базофилы, абс.',
]
```

```python
parsed_list = page1text.split('\n')
```

```python
parsed_list[85] # .split('< ')[1] #.split(', ')[0]
#.split(' ')[0] #.split('< ')[1] #.replace(', %', '') #.strip() #.split(' ')[4]
```

```python
result = []

for name in parse_list:
    i = parsed_list.index(name)
    
    #print(i)
    
    #if i != 43 and i != 62 and i != 85:
    if name.split(' ')[0] != 'RDW' and name != 'Нейтрофилы (общ.число), ' and name != 'Базофилы, %':
        result.append(
            [
                parsed_list[i],
                parsed_list[i+1].replace('*', ''),
                parsed_list[i+3].split(' - ')[0],
                parsed_list[i+3].split(' - ')[1]
            ]  
        )
    elif name.split(' ')[0] == 'RDW':
        result.append(
            [
                ' '.join(parsed_list[i].split(' ')[:4]),
                parsed_list[i].split(' ')[4].replace('*', ''),
                parsed_list[i+2].split(' - ')[0],
                parsed_list[i+2].split(' - ')[1]
            ]
        )
    elif name == 'Нейтрофилы (общ.число), ':
        result.append(
            [
                parsed_list[i],
                parsed_list[i+2].replace('*', ''),
                parsed_list[i+4].split(' - ')[0],
                parsed_list[i+4].split(' - ')[1]
            ]
        )
    elif name == 'Базофилы, %':
        print(i)
        result.append(
            [
                parsed_list[i].replace(', %', ''),
                parsed_list[i+1].replace('*', ''),
                '-1000',
                parsed_list[i+3].split('< ')[1]
            ]
        )
    #else:
    #    pass

    
address = parsed_list[5]
fio = parsed_list[6]
sex = parsed_list[8]
age = parsed_list[10].split(' ')[0]
inz = parsed_list[12]
take_date = parsed_list[14].split(' ')[0]
```

```python
df_res = pd.DataFrame(result, columns=['test_name', 'test_result', 'test_lower', 'test_upper'])

df_res['address'] = address
df_res['fio'] = fio
df_res['sex'] = sex
df_res['age'] = age
df_res['inz'] = inz
df_res['take_date'] = take_date
```

```python
df_res
```

```python
df_res['test_result'] = df_res['test_result'].astype(float)
df_res['test_lower'] = df_res['test_lower'].astype(float)
df_res['test_upper'] = df_res['test_upper'].astype(float)


df_res['test'] = df_res.apply(lambda x: (x['test_result'] < x['test_lower'] and 'lower') 
                      or (x['test_result'] <= x['test_upper'] and 'good')
                      or (x['test_result'] > x['test_upper'] and 'upper')
                     , axis=1)
```

```python
df_govno = pd.DataFrame()

for govno in df_res['test_name'].unique():
    
    temp_df_0 = df_res[df_res['test_name'] == govno]
    temp_df = temp_df_0.groupby(['test_name', 'test'])[['fio']].count().reset_index().rename(columns={'fio': 'patient_quantity'})
    
    df_govno = pd.concat([df_govno, temp_df])
```
