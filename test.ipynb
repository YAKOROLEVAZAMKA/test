```python
import pandas as pd
```

```python
df = pd.read_excel('to_sort_out.xlsx', sheet_name='Data') # читаем файл
names = ['RSTV', 'EKB', 'YRSL', 'MSK', 'SPB', 'NVRS', 'TMN'] # список того, что нужно
```

```python
df['Clean'] = ''

for i in range(len(df['Insertion Order'])):
    for name in names:
        if name in df['Insertion Order'][i]:
            df['Clean'][i] = name
```
