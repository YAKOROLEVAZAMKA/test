```python
df = pd.read_excel('фейк.xlsx')
```

```python
low_value = 71
upper_value = 74

# filter bad R-values
df_filtered = df.query("R >= @low_value and R <= @upper_value")

# reset_index
df_filtered.reset_index(drop=True, inplace=True)
#df_filtered.reset_index(inplace=True)

# rename index
#df_filtered.rename(columns={'index': 'idx'}, inplace=True)
```

```python
df_filtered
```

```python
mean_values = {
    'GTM.1': 2.4,
    'GTM.5': 3
}

offset = .5 # 50%
```

```python
res = []

start = 1


for i in range(start, df_filtered.shape[0]):
    
    # stop if found first 8 values
    if len(res) == 8:
        break
        
    # if previous id + 1 == current
    if df_filtered['id'][i-1] + 1 == df_filtered['id'][i]:
        
        kostil = 0
        for k, v in mean_values.items():
            
            if df_filtered[k][i] >= (v - v * offset) and df_filtered[k][i] <= (v + v * offset) \
            and df_filtered[k][i-1] >= (v - v * offset) and df_filtered[k][i-1] <= (v + v * offset):
                kostil += 1
                
        if kostil == len(mean_values):
            res.append(i-1)
    
    else:
        res = []
    
                
```
