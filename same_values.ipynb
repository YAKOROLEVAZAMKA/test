```python
import pandas as pd
```

```python
a = [
    [1, 'a', '2020-01-01'],
    [2, 'b', '2010-01-01'],
    [1, 'a', '2020-01-02'],
    [3, 'c', '2022-01-01'],
    [1, 'a', '2020-01-03'],
    [2, 'b', '2015-01-01'],
    [1, 'c', '2022-01-01'],
    [1, 'c', '2022-05-05'],
    [1, 'c', '2022-04-04'],
    [1, 'd', '2020-01-04'],
    [1, 'a', '2022-01-02'],
    [1, 'a', '2022-01-05'],
    [1, 'a', '2022-01-07'],
]
```

```python
df = pd.DataFrame(a, columns=['patient_id', 'mkb', 'date'])
df['date'] = pd.to_datetime(df['date'])

df
```

```python
df = df.sort_values(by=['patient_id', 'date']).reset_index(drop=True)

df
```

```python
#df['podryad'] = 0
df['nomer_obrasheniya'] = 0

patient = df['patient_id'][0]
mkb = df['mkb'][0]
pat_obr = 1
df['nomer_obrasheniya'][0] = pat_obr

for i in range(1, df.shape[0]):
    curr_patient = df['patient_id'][i]
    curr_mkb = df['mkb'][i]

    if curr_patient == patient and mkb == curr_mkb:
        df['nomer_obrasheniya'][i] = pat_obr
    elif curr_patient == patient and mkb != curr_mkb:
        pat_obr += 1
    elif curr_patient != patient:
        pat_obr = 1
    
    
    if curr_patient == patient and mkb == curr_mkb:
        #df['podryad'][i] = 1
        df['nomer_obrasheniya'][i] = pat_obr
    else:
        patient = df['patient_id'][i]
        mkb = curr_mkb = df['mkb'][i]
        #df['podryad'][i] = 1
        df['nomer_obrasheniya'][i] = pat_obr
        
        

    
```
